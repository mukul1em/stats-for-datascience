```python
from scipy import stats
import numpy as np
import seaborn as sb
%matplotlib inline
```

# CALCULATING SKEWNESS

```python
heights=[165,170,171,175,175,160,162,161,180,190,155,160,170,175]
```

```python
heights
```

```python
sb.distplot(heights) #histogram plot
```

```python
stats.skew(heights) #skewness for normal curve should be -1.96 to 1.96 , for perfect normal curve is 0
```

```python
heights_skew1=[165,170,171,175,175,160,162,161,180,190,155,160,170,175,180,190,200,220,210,210,240,240]
```

```python
sb.distplot(heights_skew1) #positively skewed
```

```python
stats.skew(heights_skew1) #s=0 is perfect normal curce . this s is near to 1 but can be taken as a normal curve as it is still under the range
```
